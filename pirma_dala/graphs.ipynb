```python
import matplotlib.pyplot as plt  # Importē matplotlib moduli grafiku veidošanai
import pandas as pd  # Importē pandas moduli datu apstrādei
from mpl_toolkits.mplot3d import Axes3D  # Importē 3D grafiku veidošanas rīku

# Ielādē CSV failus
normal_file = r"\chosen_df\50hz\iis2dh_acc_normal.csv"  # Faila ceļš ar normālu stāvokli
disbalance_file = r"\chosen_df\50hz\iis2dh_acc_disbalance.csv"  # Faila ceļš ar disbalansa stāvokli

# Nolasa datus DataFrame formātā
df_normal = pd.read_csv(normal_file)  # Ielasa pirmo failu (normāls stāvoklis)
df_disbalance = pd.read_csv(disbalance_file)  # Ielasa otro failu (disbalansa stāvoklis)

# Izvelk datus normālam gadījumam
Ax_normal = df_normal["A_x [g]"]  # Paātrinājums x asī
Ay_normal = df_normal["A_y [g]"]  # Paātrinājums y asī
Az_normal = df_normal["A_z [g]"]  # Paātrinājums z asī

# Izvelk datus disbalansa gadījumam
Ax_disbalance = df_disbalance["A_x [g]"]  # Paātrinājums x asī
Ay_disbalance = df_disbalance["A_y [g]"]  # Paātrinājums y asī
Az_disbalance = df_disbalance["A_z [g]"]  # Paātrinājums z asī

# Definē asi robežas
x_limits = (-0.1, 0.15)  # x ass robežas
y_limits = (0.5, 0.90)  # y ass robežas
z_limits = (0.4, 1)  # z ass robežas

# Zīmē normālo gadījumu
fig = plt.figure(figsize=(18, 6))  # Pielāgo figūras izmēru labākai izkārtojuma telpai
ax1 = fig.add_subplot(131, projection='3d')  # Pievieno pirmo 3D apakšgrafiku
ax1.scatter(Ax_normal, Ay_normal, Az_normal, color="green", alpha=0.6, s=20)  # Zīmē normālo stāvokli (zaļš)
ax1.set_xlim(x_limits)  # Uzstāda x ass robežas
ax1.set_ylim(y_limits)  # Uzstāda y ass robežas
ax1.set_zlim(z_limits)  # Uzstāda z ass robežas
ax1.legend(['Normal'], loc='upper left', fontsize=12)  # Parāda leģendu
ax1.grid(True)  # Ieslēdz režģi

# Zīmē disbalansa gadījumu
ax2 = fig.add_subplot(132, projection='3d')  # Pievieno otro 3D apakšgrafiku
ax2.scatter(Ax_disbalance, Ay_disbalance, Az_disbalance, color="red", alpha=0.6, s=20)  # Zīmē disbalansa stāvokli (sarkans)
ax2.set_xlim(x_limits)  # Uzstāda x ass robežas
ax2.set_ylim(y_limits)  # Uzstāda y ass robežas
ax2.set_zlim(z_limits)  # Uzstāda z ass robežas
ax2.legend(['Disbalance'], loc='upper left', fontsize=12)  # Parāda leģendu
ax2.grid(True)  # Ieslēdz režģi

# Zīmē abu gadījumu datus uz viena grafika
ax3 = fig.add_subplot(133, projection='3d')  # Pievieno trešo 3D apakšgrafiku
ax3.scatter(Ax_normal, Ay_normal, Az_normal, color="green", alpha=0.6, s=20, label="Normal")  # Zīmē normālo stāvokli (zaļš)
ax3.scatter(Ax_disbalance, Ay_disbalance, Az_disbalance, color="red", alpha=0.6, s=20, label="Disbalance")  # Zīmē disbalansa stāvokli (sarkans)
ax3.set_xlim(x_limits)  # Uzstāda x ass robežas
ax3.set_ylim(y_limits)  # Uzstāda y ass robežas
ax3.set_zlim(z_limits)  # Uzstāda z ass robežas
ax3.legend(fontsize=12)  # Parāda leģendu ar abu datu kopām
ax3.grid(True)  # Ieslēdz režģi

# Pielāgo izkārtojumu, lai grafiki nepārklātos
plt.tight_layout()

# Attēlo grafikus
plt.show()
```

```python
import pandas as pd  # Importē pandas datu apstrādei
import seaborn as sns  # Importē seaborn datu vizualizācijai
import matplotlib.pyplot as plt  # Importē matplotlib grafiku veidošanai

# Ielādē akselerometra datus no CSV failiem
acc_normal_file_path = r'\chosen_df\10hz\iis2dh_acc_normal.csv'  # Normāla stāvokļa datu faila ceļš
acc_normal_data = pd.read_csv(acc_normal_file_path)  # Ielasa normālos datus DataFrame formātā

acc_disbalance_file_path = r'\chosen_df\10hz\iis2dh_acc_disbalance.csv'  # Disbalansa stāvokļa datu faila ceļš
acc_disbalance_data = pd.read_csv(acc_disbalance_file_path)  # Ielasa disbalansa datus DataFrame formātā

# Pievieno kolonnu katram DataFrame, lai norādītu datu veidu
acc_normal_data['Type'] = 'Normal'  # Pievieno kolonnu "Type" ar vērtību "Normal"
acc_disbalance_data['Type'] = 'Disbalance'  # Pievieno kolonnu "Type" ar vērtību "Disbalance"

# Apvieno abus DataFrame vienā kopīgā DataFrame
combined_data = pd.concat([acc_normal_data, acc_disbalance_data], ignore_index=True)  # Apvieno datus un ignorē vecos indeksus

# Izveido KDE grafiku (kodolblīvuma aplēsi) A_y [g] sadalījumam
plt.figure(figsize=(8, 6))  # Definē grafika izmērus
kde_plot = sns.kdeplot(
    data=combined_data, 
    x='A_y [g]', 
    hue='Type', 
    palette={'Normal': 'green', 'Disbalance': 'red'}, 
    fill=True  # Aizpilda grafiku ar krāsām pēc tipa
)

# Noņem konkrētas etiķetes (x un y asīm)
kde_plot.set_xlabel('')  # Noņem x ass etiķeti
kde_plot.set_ylabel('')  # Noņem y ass etiķeti

# Attēlo grafiku
plt.show()
```

#MICROPHONE SPECTROGRAM

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def create_spectrogram(data, sampling_rate, window_size=256, overlap=128):
    """
    Izveido spektrogrammu no audio datiem.

    Parametri:
        data (ndarray): Audio dati.
        sampling_rate (int): Audio datu paraugu ņemšanas frekvence.
        window_size (int): FFT loga izmērs.
        overlap (int): Pārklājums starp secīgiem logiem.

    Atgriež:
        ndarray: Spektrogrammas dati.
        ndarray: Frekvences.
        ndarray: Laika intervāli.
    """
    num_samples = len(data)  # Datu paraugu skaits
    num_overlap = window_size - overlap  # Pārklājuma izmērs
    num_windows = (num_samples - window_size) // num_overlap + 1  # Logu skaits

    frequencies, time_intervals, spectrogram = \
        plt.specgram(data, NFFT=window_size, Fs=sampling_rate,
                     noverlap=overlap, scale='dB')

    return spectrogram, frequencies, time_intervals

# Ielādē audio datus no CSV faila
audio_data = pd.read_csv(r'\imp23absu_mic.csv')
sampling_rate = 48000  # Paraugu ņemšanas frekvence
audio_data  # Parāda ielādētos datus

audio_data = audio_data['MIC']  # Izvēlas kolonnu 'MIC' no datiem

# Izveido spektrogrammu
spectrogram, frequencies, time_intervals = \
    create_spectrogram(audio_data, sampling_rate)

# Zīmē spektrogrammu
plt.figure(figsize=(10, 5))  # Uzstāda figūras izmēru
plt.imshow(spectrogram, aspect='auto', origin='lower',
            extent=[time_intervals.min(), time_intervals.max(),
                    frequencies.min(), frequencies.max()])
plt.xlabel('Laiks (s)')  # X ass nosaukums
plt.ylabel('Frekvence (Hz)')  # Y ass nosaukums
plt.title('Spektrogramma')  # Grafika nosaukums
plt.colorbar(label='Intensitāte (dB)')  # Krāsu joslas nosaukums
plt.show()  # Parāda izveidoto grafiku
```
